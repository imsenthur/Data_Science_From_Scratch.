```python
%matplotlib inline
from matplotlib import pyplot as plt

x = [1,2,3,4,5,6,7,8,9]
y = [5,6,7,8,9,1,2,3,5]

plt.plot(x, y, color='green', marker='o', linestyle = 'solid')
plt.title("TestGraph")
plt.ylabel("TestYs")
plt.show()
```

```python
'''BARCHARTS'''
habbits = ["Sleeping", "Eating", "Having fun", "Working", "Chilling"]
hours = [6, 2, 4, 10, 2]
xs = [i + 0.1 for i, _ in enumerate(habbits)]
plt.bar(xs, hours)

plt.xticks([i+0.1 for i, _ in enumerate(habbits)], habbits)
plt.ylabel("Hours")
plt.title("Routine")
plt.show()
```

```python
'''LINECHARTS'''
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
squares = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x+y for x,y in zip(variance, squares)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'r-', label = 'Variance')
plt.plot(xs, squares, 'g-.', label = 'Squares')
plt.plot(xs, total_error, 'b:', label = 'total_error')

plt.legend(loc=0)
plt.ylabel("Model complexity")
plt.title("Bias_Variance Tradeoff")
```

```python
'''SCATTERPLOTS'''
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)
for minute_count, friend_count, label in zip(minutes, friends, labels):
    plt.annotate(label, xy = (friend_count, minute_count), xytext = (5, -5), textcoords = 'offset points' )

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```
